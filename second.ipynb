```python
import numpy as np
import scipy as sp
from math import sin
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python
trace = [] # 추적

def f(x):
    value = x ** 2 + 10 * np.sin(x)
    trace.append((x,value))
    return value

X = [-8 + 18 * i / 9999 for i in range(1000)]
Y = [f(x) for x in X]

plt.style.use("dark_background")
plt.figure()
plt.plot(X,Y,'w-')
for start, color in [(8,'red') , (-6,'yellow')]:
    trace = []
    solution = minimize(f, [start], method = 'nelder-mead')
    x,y = solution.x[0] , solution.fun
    plt.plot(*zip(*trace), '.-', c =color, label=f'start: {start:.1f}') # x: -6부터 시작하는 곳에서 최적화, x:8부터 시작하는 곳에서 부터 최적화
    plt.plot(*trace[0], 'o' , c= color) # 최적화 시작점
    plt.text(x,-23,f'{x:.3f}',c=color,ha='center') # 미니멈이 되는 y값
    plt.plot([x,x],[-18,y], '--', c=color) # 
    print(f'x: {x} , y: {y}')
    print(f'trace : {trace}')
    
plt.xticks(range(-5,15,5))
plt.yticks(range(-25,100,25))
plt.minorticks_on()
plt.legend()
plt.show()
```
